```python
#19. Normalize [10,20,30,40,50] using Min-Max normalization.
arr = np.array([10,20,30,40,50])
a_min = arr.min()
a_max = arr.max()
norm_data = (arr-a_min)/(a_max-a_min)
norm_data
```

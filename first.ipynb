```python
import numpy as np 
import time

n = 1000

list1 = list(range(n))
list2 = list(range(n))

start = time.time()
result_list = [list1[i] + list2[i] for i in range(n)]
list_time = time.time() - start

arr1 = np.array(n)
arr2 = np.array(n)

start = time.time()
result_arr = arr1 + arr2
numpy_time = time.time() - start

print(f"List Time : {list_time:.6f} sec")
print(f"Numpy Time : {numpy_time:.6f} sec")
print(f"Numpy is approximately  {list_time/numpy_time:.6f}X faster")
```

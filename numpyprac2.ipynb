```python
#Create a Chackerboard pattern 
import numpy as np
arr = np.arange(1,65).reshape(8,8)
for i in range(0,8):
    if i%2==0:
        arr[i,::2]=0
    else:
        arr[i,1::2]=0
for i in range(0,8):
    if i%2==1:
        arr[i,::2]=1
    else:
        arr[i,1::2]=1

print(arr)
```

```python
import numpy as np
arr = np.ones((8,8))
for i in range(0,8):
    if i%2==0:
        arr[i,::2]=0
    else:
        arr[i,1::2]=0

print(arr)
```

```python
arr = np.zeros((8,8))
arr[1::2,::2] = 1
arr[::2,1::2] = 1
print(arr)
```
